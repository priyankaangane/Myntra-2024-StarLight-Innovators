```python
file_path = '/content/drive/My Drive/myntra.xlsx'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
```

```python
# Load the Excel file
data = pd.read_excel(file_path)
```

```python
# Display the first few rows to check the data
print(data.head())
```

```python
# Step 3: Perform the Analysis
# Combine likes and comments into a new column
data['Total Engagement'] = data['Likes'] + data['Comments']
```

```python
# Sort the data by Total Engagement in descending order
sorted_data = data.sort_values(by='Total Engagement', ascending=False)
```

```python
# Display the first few rows of the sorted data
print(sorted_data.head())
```

```python
# Plot the distribution of Total Engagement
plt.figure(figsize=(14, 7))
sns.histplot(sorted_data['Total Engagement'], kde=True)
plt.title('Distribution of Total Engagement (Likes + Comments)')
plt.xlabel('Total Engagement')
plt.ylabel('Frequency')
plt.show()
```

```python
# Combine all hashtags into a single list
all_hashtags = ' '.join(data['Hashtags']).split()
# Clean hashtags by removing special characters and converting to lower case
all_hashtags = [re.sub(r'\W+', '', tag).lower() for tag in all_hashtags]
```

```python
# Count the frequency of each hashtag
hashtag_counts = Counter(all_hashtags)
most_common_hashtags = hashtag_counts.most_common(10)
```

```python
# Convert to DataFrame for plotting
hashtag_df = pd.DataFrame(most_common_hashtags, columns=['hashtag', 'count'])
```

```python
# Plot the most common hashtags
plt.figure(figsize=(14, 7))
sns.barplot(x='hashtag', y='count', data=hashtag_df)
plt.title('Top 10 Most Common Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 4: Save the results to a new Excel file
sorted_file_path = '/content/drive/My Drive/sorted_myntra_posts.xlsx'
sorted_data.to_excel(sorted_file_path, index=False)
```
